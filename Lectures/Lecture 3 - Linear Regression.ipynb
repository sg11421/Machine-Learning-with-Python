## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generating Data

```python
rng = np.random.RandomState(10)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y)
m=x.shape[0]
```

## Cost fn

```python
def cost(y_hat, y):
    l = (y_hat - y) ** 2 / 2
    cost = l.mean()
    return cost
```

## Hypothesis

```python
def hypothesis(x,w,b):
    return w*x+b
```

## Gradient Descent

```python
def gradient_descent(x,y,w,b,learning_rate,m):
   y_hat=hypothesis(x,w,b)
   dw = (2.0/m) * np.sum(np.multiply(x,(y_hat-y)))
   db = (2.0/m) * np.sum(y_hat-y)
   w = w -learning_rate * dw
   b = b - learning_rate * db
   return w,b
```

## Implementation

```python
# Random Initialisation
w = np.random.randn(1)
b = np.random.randn(1)

# Hyperparameters
learning_rate = 0.00005
num_epochs = 100000

L=[]
for i in range(num_epochs):
    w, b = gradient_descent(x,y,w,b,learning_rate,m)
    y_hat = hypothesis(x,w,b)
    L.append(cost(y_hat,y))
    if (i%(num_epochs/10))==0:
        plt.plot(x,y,'bo')
        plt.plot(x,y_hat,'g--')
        plt.show()
```

## Results

```python
print(f"Obtained value of w: {w}")
print(f"Obtained value of b: {b}")
print(f"Obtained line: y={w}x+{b}")
plt.plot(x,y,'bo')
plt.plot(x,y_hat,'g--')
plt.show()
```

## Plot of cost fn vs no of iterations

```python
plt.plot(L,'r')
plt.show()
```

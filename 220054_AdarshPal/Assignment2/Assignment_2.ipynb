```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
X = np.random.rand(100)
X
```

```python
y = 4 + 5*X + .7*np.random.randn(100)
y
```

```python
print(X.shape)
print(y.shape)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
m = X.shape[0]
m
```

```python
# defining hypothesis
def hypothesis(w,X,b):
  return w*X + b
```

```python
#defining cost function
def cost(y_hat,y):
  c = abs((y_hat - y)**3)/3

  return c.mean()
```

```python
def gradient_descent(X, y, w, b, lr, m):
    y_hat = hypothesis(w, X, b)
    dw = (1.0 / m) * np.sum(np.multiply(-X, (abs(y_hat - y))*(y-y_hat))) ## on taking derivative the 3 gets cancelled out hence numerator becomes 1.0 ##
    db = (-1.0 / m) * np.sum((abs(y_hat - y))*(y-y_hat))


    w = w - lr * dw
    b = b - lr * db
    return w, b
```

```python
w = np.random.randn(1)
b = np.random.randn(1)
# Hyperparameters
lr = 0.0999
num_epochs = 3000
L = []


for i in range(num_epochs):
    w, b = gradient_descent(X,y,w,b,lr,m)
    y_hat = hypothesis(w,X,b)
    L.append(cost(y_hat,y))
    print(f"w is {w}, b is {b}, cost is{cost(y_hat,y)}")
```

```python
print(f"w is {w}, b is {b}")
```

```python
plt.scatter(X,y)
plt.plot(X,w*X + b, 'r')
```

```python
plt.plot(L,'g')
```

```python
### POLYNOMIAL REGRESSION ###
```

```python
def hypothesis_(w1,w2,b_,X):
  return w1*X**2 + w2*X + b_
```

```python
m_ = X.shape[0]
m_
```

```python
def cost_(y,yn_hat):
  c = (abs(y-yn_hat)**7)/7
  return c.mean()
```

```python
def gradient_descent_(X, y, w1,w2, b_, lr_, m_):
    yn_hat = hypothesis_(w1,w2,b_,X)

    dw1 = (1.0 / m_) * np.sum(np.multiply(-X**2, (abs(y-y_hat))*(y - yn_hat)**5)) ## on taking derivative the 7 gets cancelled out hence numerator becomes 1.0 ##
    dw2 = (-1.0 / m_) * np.sum(np.multiply(X, (abs(y-yn_hat))*(y - yn_hat)**5))
    db = (-1.0 / m_) * np.sum((abs(y-yn_hat))*((y - yn_hat)**5))
    w1 = w1 - lr_ * dw1
    w2 = w2 - lr_ * dw2
    b_ = b_ - lr_ * db
    return w1,w2, b
```

```python
w1 = np.random.randn(1)
w2 = np.random.randn(1)
b_ = np.random.randn(1)
# Hyperparameters
lr_ = 0.00001
num_epochs_ = 14000
L_ = []


for j in range(num_epochs_):
    w1,w2, b_ = gradient_descent_(X,y,w1,w2,b_,lr_,m_)
    yn_hat = hypothesis_(w1,w2,b_,X)
    L_.append(cost_(y,yn_hat))
    print(f"w1 is {w1},w2 is {w2} b_ is {b_}, cost is {cost_(yn_hat,y)}")
```

```python
plt.scatter(X,y)
plt.plot(X,w1*X**2 + w2*X + b_ ,'r')
```

```python
plt.plot(L_,'r')
```

```python
def model(xf,yf,lr,iter):
  m = xf.shape[0]
  theta = np.ones((xf.shape[1],1))
  L=[]
  for i in range(iter):
    y_pred = np.dot(xf,theta)
    cost = (1/(2*m))*np.sum(np.square(y_pred-yf))
    d_theta = (1/m)*np.dot(xf.T,(y_pred-yf))
    theta= theta -  lr*d_theta
    L.append(cost)
    print(f"cost is {cost}")

  return theta, L
```

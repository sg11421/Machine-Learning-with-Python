### Q1. Matrix Multiplication

##### (i) Check if matrix multiplication is valid

```python
def isValid(A,B):
   if A.shape[1] != B.shape[0]:
    return False
   else:
    return True
```

##### (ii) Using loops (without using numpy)

```python
def matrix_multiply(A,B):
  result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]


  for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]

  return result
```

##### (iii) Using numpy

```python
def matrix_multiply_2(A,B):
  Result = np.dot(A,B)
  return(Result)
```

### Q2. Z-Score Normalisation

Z-score normalization refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

New value = (x – μ) / σ

where:

x: Original value

μ: Mean of data

σ: Standard deviation of data

##### (i) Without using numpy

```python
def mean(x):
    mean_value = sum(x) / len(x)
    return mean_value
```

```python
def standard_deviation(x):
  mean_value = sum(x) / len(x)
  variance = sum((xi - mean_value) ** 2 for xi in x) / len(x)
  std_dev = variance ** 0.5
  return std_dev
```

```python
def zscore_normalisation(x):
 mean_value = sum(x) / len(x)
 variance = sum((xi - mean_value) ** 2 for xi in x) / len(x)
 std_dev = variance ** 0.5
 normalized_x = [(xi - mean_value) / std_dev for xi in x]
 return normalized_x
```

##### (ii) Using numpy

Numpy has in_built functions for calculating mean and standard deviation

```python
def zscore_normalisation_2(x):
  mean_val=np.mean(x)
  std_val=np.std(x)
  zscore_normalised_2=(x-mean_val)/std_val 
  return zscore_normalised_2
```

### Q3. Sigmoid fn and its derivative

The sigmoid function is a mathematical function that maps any input value to a value between 0 and 1.

It is defined mathematically as s(x) = 1/(1+e^(-x)).

##### (i) Write a fn to implement sigmoid fn

```python
def sigmoidfn(x):
   return 1 / (1 + np.exp(-x))
```

##### (ii) Write a fn to implement derivative of sigmoid fn

```python
def derivative(x):
  return np.exp(-x) / (1 + np.exp(-x)) ** 2
```

# Section 3: Plotting with Matplotlib

### Plot the properties (specific volume, specific internal energy, specific enthalpy, specific entropy) vs Pressure for saturated liquid.

Note:
- Try using the subplot feature of matplotlib(Explore it!!)
- Provide appropriate title, labels, markersize and other parameters to the plot

```python
Specific_Volume=df.loc[(df['Property']=='V'),'Liq_Sat'].values
P_V=df.loc[(df['Property']=='V'),'Pressure'].values

Specific_InternalEnergy=df.loc[(df['Property']=='U'),'Liq_Sat'].values
P_U=df.loc[(df['Property']=='U'),'Pressure'].values

Specific_Enthalpy=df.loc[(df['Property']=='H'),'Liq_Sat'].values
P_H=df.loc[(df['Property']=='H'),'Pressure'].values

Specific_Entropy=df.loc[(df['Property']=='S'),'Liq_Sat'].values
P_S=df.loc[(df['Property']=='S'),'Pressure'].values

fig, axs=plt.subplots(2, 2, figsize=(10,10))

axs[0,0].plot(P_V,Specific_Volume,color='b', marker='X', linestyle='--',linewidth=1, markersize=3)
axs[0,0].set_title("Specific Volume vs Pressure")
axs[0,0].set_xlabel("Pressure(kPa)")
axs[0,0].set_ylabel("Specific Volume")

axs[0,1].plot(P_U,Specific_InternalEnergy,color='r', marker='*', linestyle='--',linewidth=1, markersize=3)
axs[0,1].set_title("Specific Internal Energy vs Pressure")
axs[0,1].set_xlabel("Pressure(kPa)")
axs[0,1].set_ylabel("Specific Internal Energy")

axs[1,0].plot(P_H,Specific_Enthalpy,color='g', marker='D', linestyle='--',linewidth=1, markersize=3)
axs[1,0].set_title("Specific Enthalpy vs Pressure")
axs[1,0].set_xlabel("Pressure(kPa)")
axs[1,0].set_ylabel("Specific Enthalpy")

axs[1,1].plot(P_S,Specific_Entropy,color='purple', marker='o', linestyle='--',linewidth=1, markersize=3)
axs[1,1].set_title("Specific Entropy vs Pressure")
axs[1,1].set_xlabel("Pressure(kPa)")
axs[1,1].set_ylabel("Specific Entropy")

plt.tight_layout()
plt.show()
```

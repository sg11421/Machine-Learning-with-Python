```python
def isValid(A,B):
  if len(A[0]) != len(B):
    return False
  else:
    return True
```

##### (ii) Using loops (without using numpy)

```python
def matrix_multiply(A,B):
  result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]

    # Iterate through rows of A
  for i in range(len(A)):
        # Iterate through columns of B
        for j in range(len(B[0])):
            # Iterate through rows of B
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]

  return result
```

### Q2. Z-Score Normalisation

Z-score normalization refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

New value = (x – μ) / σ

where:

x: Original value

μ: Mean of data

σ: Standard deviation of data

##### (i) Without using numpy

```python
def mean(x):
  mean_value = sum(x) / len(x)
  return mean_value
```

```python
def standard_deviation(x):
  mean_value = sum(x) / len(x)
  variance = sum((xi - mean_value) ** 2 for xi in x) / len(x)
  std_dev = variance ** 0.5
  return std_dev
```

```python
def zscore_normalisation(x):
 mean_value = sum(x) / len(x)
 variance = sum((xi - mean_value) ** 2 for xi in x) / len(x)
 std_dev = variance ** 0.5


 normalized_x = [(xi - mean_value) / std_dev for xi in x]

 return normalized_x
```

##### (ii) Using numpy

Numpy has in_built functions for calculating mean and standard deviation

```python
def zscore_normalisation_2(x):
 mean_value = np.mean(x)
 std_dev = np.std(x)


 normalized_x = (x - mean_value) / std_dev

 return normalized_x
```

### Q3. Sigmoid fn and its derivative

The sigmoid function is a mathematical function that maps any input value to a value between 0 and 1.

It is defined mathematically as s(x) = 1/(1+e^(-x)).

```python
def sigmoidfn(x):
   return 1 / (1 + np.exp(-x))
```

# Section 3: Plotting with Matplotlib

### Plot the properties (specific volume, specific internal energy, specific enthalpy, specific entropy) vs Pressure for saturated liquid.

Note:
- Try using the subplot feature of matplotlib(Explore it!!)
- Provide appropriate title, labels, markersize and other parameters to the plot

```python


# Extract data for saturated liquid
sat_liquid_df = df[['Pressure', 'Liq_Sat']]

# Plotting
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Specific Volume vs Pressure
axs[0, 0].plot(sat_liquid_df['Pressure'], sat_liquid_df['Liq_Sat'], marker='o', markersize=6, linestyle='-', color='b')
axs[0, 0].set_title('Specific Volume vs Pressure for Saturated Liquid')
axs[0, 0].set_xlabel('Pressure (kPa)')
axs[0, 0].set_ylabel('Specific Volume')
axs[0, 0].grid(True)

# Specific Internal Energy vs Pressure
axs[0, 1].plot(sat_liquid_df['Pressure'], sat_liquid_df['Liq_Sat'], marker='o', markersize=6, linestyle='-', color='r')
axs[0, 1].set_title('Specific Internal Energy vs Pressure for Saturated Liquid')
axs[0, 1].set_xlabel('Pressure (kPa)')
axs[0, 1].set_ylabel('Specific Internal Energy')
axs[0, 1].grid(True)

# Specific Enthalpy vs Pressure
axs[1, 0].plot(sat_liquid_df['Pressure'], sat_liquid_df['Liq_Sat'], marker='o', markersize=6, linestyle='-', color='g')
axs[1, 0].set_title('Specific Enthalpy vs Pressure for Saturated Liquid')
axs[1, 0].set_xlabel('Pressure (kPa)')
axs[1, 0].set_ylabel('Specific Enthalpy')
axs[1, 0].grid(True)

# Specific Entropy vs Pressure
axs[1, 1].plot(sat_liquid_df['Pressure'], sat_liquid_df['Liq_Sat'], marker='o', markersize=6, linestyle='-', color='purple')
axs[1, 1].set_title('Specific Entropy vs Pressure for Saturated Liquid')
axs[1, 1].set_xlabel('Pressure (kPa)')
axs[1, 1].set_ylabel('Specific Entropy')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

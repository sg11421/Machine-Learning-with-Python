### Q1. Matrix Multiplication

##### (i) Check if matrix multiplication is valid

```python
def isValid(A,B):
  #your code here
  if A.shape[1]==B.shape[0]:
    return 1
  return -1
```

##### (ii) Using loops (without using numpy)

```python
def matrix_multiply(A,B):
  #your code here
  n1=A.shape[1]
  m1=A.shape[0]
  m2=B.shape[1]
  new_matrix=np.ones((m1,m2))
  for i in range(m1):
    for k in range(m2):
      sum1=0
      for j in range(n1):
        sum1+=A[i,j]*B[j,k]
      new_matrix[i,k]=sum1
  return new_matrix
```

##### (iii) Using numpy

```python
def matrix_multiply_2(A,B):
  #your code here
  return A.dot(B)
```

### Q2. Z-Score Normalisation

Z-score normalization refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

New value = (x – μ) / σ

where:

x: Original value

μ: Mean of data

σ: Standard deviation of data

##### (i) Without using numpy

```python
def mean(x):
  #your code here
  sum2=0
  for i in range(len(x)):
    sum2+=x[i]
  return sum2/(len(x))
```

```python
def standard_deviation(x):
  #your code here
  sum3=0
  for j in range(len(x)):
    sum3+=(x[j]-mean(x))**2
  return np.sqrt(sum3/len(x)) 
```

```python
def zscore_normalisation(x):
  #your code here
  mean_used=mean(x)
  std_used=standard_deviation(x)
  for k in range(len(x)):
   x[k] =(x[k]- mean_used)/std_used
  print(x)  
```

##### (ii) Using numpy

Numpy has in_built functions for calculating mean and standard deviation

```python
def zscore_normalisation_2(x):
  #your code here
  mean=np.mean(x)
  std_dev=np.std(x)
  print((x-mean)/std_dev)  
```

### Q3. Sigmoid fn and its derivative

The sigmoid function is a mathematical function that maps any input value to a value between 0 and 1.

It is defined mathematically as s(x) = 1/(1+e^(-x)).

##### (i) Write a fn to implement sigmoid fn

```python
def sigmoidfn(x):
  #your code here
  return 1/(1+np.exp(-x)) 
```

##### (ii) Write a fn to implement derivative of sigmoid fn

```python
def derivative(x):
  #your code here
  return sigmoidfn(x)*(1-sigmoidfn(x))  
```

Note:
- Try using the subplot feature of matplotlib(Explore it!!)
- Provide appropriate title, labels, markersize and other parameters to the plot

```python
#your code here
fig, axs = ptl.subplots(nrows=2, ncols=2, figsize=(10, 12))
# Plot 'H' versus 'Pressure'
axs[0,0].plot(H_data['Pressure'], H_data['Liq_Sat'],"go" )
axs[0,0].set_xlabel('Pressure')
axs[0,0].set_ylabel('H')
axs[0,0].set_title("Specific Enthalpy(H) vs Pressure")

# Plot 'V' versus 'Pressure'
axs[1,0].plot(V_data['Pressure'], V_data['Liq_Sat'], "bo")
axs[1,0].set_xlabel('Pressure')
axs[1,0].set_ylabel('V')
axs[1,0].set_title("Specific Volume(V) vs Pressure")

# Plot 'U' versus 'Pressure'
axs[0,1].plot(U_data['Pressure'], U_data['Liq_Sat'], "ro")
axs[0,1].set_xlabel('Pressure')
axs[0,1].set_ylabel('U')
axs[0,1].set_title("Specific Internal Energy(U) vs Pressure")

# Plot 'S' versus 'Pressure'
axs[1,1].plot(S_data['Pressure'], S_data['Liq_Sat'], "yo")
axs[1,1].set_xlabel('Pressure')
axs[1,1].set_ylabel('S')
axs[1,1].set_title("Specific Entropy(S) vs Pressure")
ptl.suptitle("H,U,V and S Vs 'Pressure' for 'sat_liq '")
ptl.tight_layout()
ptl.show()
```

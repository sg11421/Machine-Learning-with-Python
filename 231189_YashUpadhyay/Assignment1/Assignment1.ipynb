### Q1. Matrix Multiplication

##### (i) Check if matrix multiplication is valid

```python
def isValid(A,B):
  #your code here
    row1,column1 = np.shape(A)
    row2,column2 = np.shape(B)
    if(column1==row2):
        return True
    else:
        return False
    
```

##### (ii) Using loops (without using numpy)

```python
def matrix_multiply(A,B):
  #your code here
    c=[([0]*len(B[0])) for x in range(len(A))]
    for row in range(len(A)):
        for column in range(len(B[0])):
            for i in range(len(B)):
                c[row][column]+=A[row][i]*B[i][column]
    return c         
```

##### (iii) Using numpy

```python
def matrix_multiply_2(A,B):
  #your code here
    return np.dot(A,B)
```

### Q2. Z-Score Normalisation

Z-score normalization refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

New value = (x – μ) / σ

where:

x: Original value

μ: Mean of data

σ: Standard deviation of data

##### (i) Without using numpy

```python
def mean(x):
  #your code here
    a=0.0
    b=0.0
    for index in range(len(x)):
        a+=x[index]
    b=a/len(x)
    return b
```

```python
def standard_deviation(x):
  #your code here
    d=0.0
    c=0.0
    for index in range(len(x)):
        d=d+((x[index]-mean(x))**2)
    c=((d/len(x))**0.5)
    return c
```

```python
def zscore_normalisation(x):
  #your code here
    y=x.copy()
    m=mean(x)
    n=standard_deviation(x)
    for index in range(len(x)):
        y[index]=((x[index]-m)/n)
    return y
    
```

##### (ii) Using numpy

Numpy has in_built functions for calculating mean and standard deviation

```python
def zscore_normalisation_2(x):
  #your code here
    a=np.mean(x)
    b=np.std(x)
    y=x
    for index in range(len(x)):
        y[index]=(x[index]-a)/b
    return y
```

### Q3. Sigmoid fn and its derivative

The sigmoid function is a mathematical function that maps any input value to a value between 0 and 1.

It is defined mathematically as s(x) = 1/(1+e^(-x)).

##### (i) Write a fn to implement sigmoid fn

```python
def sigmoidfn(x):
  #your code here
    y=np.zeros((len(x),len(x[0])))
    for rows in range(len(x)):
        for column in range(len(x[0])):
            y[rows][column]=(1/(1+(np.exp(-x[rows][column]))))
    return y
            
```

##### (ii) Write a fn to implement derivative of sigmoid fn

```python
def derivative(x):
  #your code here
    y=np.zeros((len(x),len(x[0])))
    for rows in range(len(x)):
        for column in range(len(x[0])):
            y[rows][column]=((np.exp(-x[rows][column]))*((1+(np.exp(-x[rows][column])))**(-2)))
    return y
```

### Q1. Matrix Multiplication

##### (i) Check if matrix multiplication is valid

```python
def isValid(A,B):
  return A.shape[1] == B.shape[0]
```

##### (ii) Using loops (without using numpy)

```python
def matrix_multiply(A,B):
  rows_A, cols_A = len(A), len(A[0])
  rows_B, cols_B = len(B), len(B[0])
  result_matrix = [[0 for _ in range(cols_B)] for _ in range(rows_A)]
  for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                result_matrix[i][j] += A[i][k] * B[k][j]

  return result_matrix
```

##### (iii) Using numpy

```python
def matrix_multiply_2(A,B):
  return np.dot(A, B)
```

### Q2. Z-Score Normalisation

Z-score normalization refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

New value = (x – μ) / σ

where:

x: Original value

μ: Mean of data

σ: Standard deviation of data

##### (i) Without using numpy

```python
def mean(x):
    total_sum = 0
    count = 0

    for value in x:
        total_sum += value
        count += 1

    mean = total_sum / count
    return mean
```

```python
def standard_deviation(x):
    mean1 = mean(x)
    sum_squared_diff = sum((value - mean1) ** 2 for value in x)
    variance = sum_squared_diff / (len(x) - 1)
    std_deviation = variance ** 0.5

    return std_deviation
```

```python
def zscore_normalisation(x):
    mean1 = mean(x)
    std = standard_deviation(x)

    NewVal = [(value - mean1) / std for value in x]

    return NewVal
    
```

##### (ii) Using numpy

Numpy has in_built functions for calculating mean and standard deviation

```python
def zscore_normalisation_2(x):
  NewVal = (x - np.mean(x)) / np.std(x)
  return NewVal
```

### Q3. Sigmoid fn and its derivative

The sigmoid function is a mathematical function that maps any input value to a value between 0 and 1.

It is defined mathematically as s(x) = 1/(1+e^(-x)).

##### (i) Write a fn to implement sigmoid fn

```python
def sigmoidfn(x):
  return 1 / (1 + np.exp(-x))
```

##### (ii) Write a fn to implement derivative of sigmoid fn

```python
def derivative(x):
  sigmoid_x = sigmoidfn(x)
  return sigmoid_x * (1 - sigmoid_x)
```

# Section 3: Plotting with Matplotlib

### Plot the properties (specific volume, specific internal energy, specific enthalpy, specific entropy) vs Pressure for saturated liquid.

Note:
- Try using the subplot feature of matplotlib(Explore it!!)
- Provide appropriate title, labels, markersize and other parameters to the plot

```python
sat_liquid_data = data[data['Property'] == 'S']

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(sat_liquid_data['Pressure'], sat_liquid_data['75'], marker='o', markersize=8, label='Specific Volume')
axs[0, 0].set_title('Specific Volume vs Pressure')
axs[0, 0].set_xlabel('Pressure')
axs[0, 0].set_ylabel('Specific Volume')

axs[0, 1].plot(sat_liquid_data['Pressure'], sat_liquid_data['100'], marker='s', markersize=8, label='Specific Internal Energy')
axs[0, 1].set_title('Specific Internal Energy vs Pressure')
axs[0, 1].set_xlabel('Pressure')
axs[0, 1].set_ylabel('Specific Internal Energy')

axs[1, 0].plot(sat_liquid_data['Pressure'], sat_liquid_data['125'], marker='^', markersize=8, label='Specific Enthalpy')
axs[1, 0].set_title('Specific Enthalpy vs Pressure')
axs[1, 0].set_xlabel('Pressure')
axs[1, 0].set_ylabel('Specific Enthalpy')

axs[1, 1].plot(sat_liquid_data['Pressure'], sat_liquid_data['150'], marker='*', markersize=8, label='Specific Entropy')
axs[1, 1].set_title('Specific Entropy vs Pressure')
axs[1, 1].set_xlabel('Pressure')
axs[1, 1].set_ylabel('Specific Entropy')

plt.show()
```

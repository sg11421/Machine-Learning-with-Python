### Q1. Matrix Multiplication

##### (i) Check if matrix multiplication is valid

```python
def isValid(A,B):
  #your code here
    shape_a = A.shape
    shape_b = B.shape
    if shape_a[1] == shape_b[0]:
        return 1
    elif shape_a[0] == shape_b[1]:
        return 1
```

##### (ii) Using loops (without using numpy)

```python
def matrix_multiply(A,B):
  #your code here
    shape_a = A.shape
    shape_b = B.shape
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    if shape_a[1] == shape_b[0]:
        # Iterate through rows of A
        for i in range(len(A)):
            # Iterate through columns of A
            for j in range(len(B[0])):
                # Iterate through rows of B
                for k in range(len(B)):
                    result[i][j] += A[i][k] * B[k][j]

        return result
    else:
        # Iterate through rows of B
        for i in range(len(B)):
            # Iterate through columns of B
            for j in range(len(A[0])):
                # Iterate through rows of A
                for k in range(len(A)):
                    result[i][j] += B[i][k] * A[k][j]

        return result
```

##### (iii) Using numpy

```python
def matrix_multiply_2(A,B):
  #your code here
    shape_a = A.shape
    shape_b = B.shape
    if shape_a[1] == shape_b[0]:
        result = np.dot(A, B)
        return result
    else:
        result = np.dot(B, A)
        return result
```

### Q2. Z-Score Normalisation

Z-score normalization refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

New value = (x – μ) / σ

where:

x: Original value

μ: Mean of data

σ: Standard deviation of data

##### (i) Without using numpy

```python
def mean(x):
  #your code here
    ans = 0
    n = len(x)
    for i in range(n):  # Iterate through the range of n
        ans = ans + x[i]
    return ans / n
```

```python
def standard_deviation(x):
  #your code here
    n = len(x)
    mew = mean(x)
    summation = 0
    for i in range(n):  # Iterate through the range of n
        summation = summation + ((x[i] - mew) ** 2)
    SD = (summation / n) ** 0.5  # Standard deviation formula
    return SD
```

```python
def zscore_normalisation(x):
    n = len(x)
    mew = mean(x)
    sd = standard_deviation(x)
    for i in range(n):  # Iterate through the range of n
        x[i] = (x[i] - mew) / sd
    return x
```

##### (ii) Using numpy

Numpy has in_built functions for calculating mean and standard deviation

```python
def zscore_normalisation_2(x):
  #your code here
    mean = np.mean(x)
    SD = np.std(x)
    for i in range(len(x)):  # Iterate through the indices of x
        x[i] = (x[i] - mean) / SD
    return x
```

### Q3. Sigmoid fn and its derivative

The sigmoid function is a mathematical function that maps any input value to a value between 0 and 1.

It is defined mathematically as s(x) = 1/(1+e^(-x)).

##### (i) Write a fn to implement sigmoid fn

```python
def sigmoidfn(x):
  #your code here
    return 1 / (1 + np.exp(-x))
```

##### (ii) Write a fn to implement derivative of sigmoid fn

```python
def derivative(x):
  #your code here
    return 1/(np.exp(-x)+np.exp(x)+2)
```

# Section 3: Plotting with Matplotlib

### Plot the properties (specific volume, specific internal energy, specific enthalpy, specific entropy) vs Pressure for saturated liquid.

Note:
- Try using the subplot feature of matplotlib(Explore it!!)
- Provide appropriate title, labels, markersize and other parameters to the plot

```python
#your code here
# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# List of separate DataFrames
dataframes = [V_data, U_data, H_data, S_data]

# List of property names
properties = ['Specific Volume', 'Specific Internal Energy', 'Specific Enthalpy', 'Specific Entropy']

# Plotting properties vs Pressure for saturated liquid
for i, (df, prop) in enumerate(zip(dataframes, properties)):
    ax = axs[i // 2, i % 2]
    ax.plot(df['Pressure'], df['Liq_Sat'], marker='o', linestyle='-', markersize=8)
    ax.set_title(f"{prop} vs Pressure for Saturated Liquid")
    ax.set_xlabel('Pressure')
    ax.set_ylabel(prop)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
